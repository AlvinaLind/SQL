## Del 1. (Grundläggande)

#### Konstruerad Tabell: user_actions med id, user_id, action, timestamp

```python
import sqlite3 
conn = sqlite3.connect("user_action.db")
cursor = conn.cursor()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS user_action (
    id INTEGER PRIMARY KEY, 
    user_id INTEGER,
    action TEXT,
    timestamp TEXT
)
""")
```

```python
actions = [(12, 'login', '2025-12-13 19:08:41'),
          (8, 'change password', '2025-12-13 19:09:02'),
          (2, 'add to favorites', '2025-12-13 19:09:55'),
          (7, 'login', '2025-12-13 19:11:21'),
          (3, 'add to cart', '2025-12-13 19:12:00'), 
          (2, 'add to cart', '2025-12-13 19:12:01'),
          (1, 'logout', '2025-12-13 19:13:37'), 
          (4, 'purchase made', '2025-12-13 10:13:37')]
cursor.executemany("""
INSERT INTO user_action (user_id, action, timestamp)
VALUES (?, ?, ?)""", actions)
conn.commit()
```

### 1.1 SELECT – Hämta data

```python
list(cursor.execute('SELECT * FROM user_action'))
```

### 1.2. WHERE – Filtrera data
#### Hämta alla rader där user_id = 12.

```python
list(cursor.execute('SELECT * FROM user_action WHERE user_id = 12'))
```

### 1.3. ORDER BY – Sortera data
#### LIMIT – Begränsa antal rader

```python
list(cursor.execute('SELECT * FROM user_action LIMIT 3'))
```
